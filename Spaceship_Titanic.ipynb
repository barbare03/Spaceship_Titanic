```python
cat_col = [col for col in data.columns if data[col].dtype == 'object']
print('Categorical columns:', cat_col)
num_col = [col for col in data.columns if data[col].dtype != 'object']
print('Numerical columns:', num_col)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

cols = ['CryoSleep', 'VIP']
for col in cols:
  data[col] = le.fit_transform(data[col])
  test[col] = le.fit_transform(test[col])
  print(f"{col} classes:", le.classes_)
```

```python
def clean(data):
  cols_to_drop =['HomePlanet', 'Cabin', 'Destination', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Name']
  data = data.drop([col for col in cols_to_drop if col in data.columns], axis=1)

  cols = ['CryoSleep', 'Age', 'VIP']
  for col in cols:
    data[col].fillna(data[col].median(), inplace=True)
  return data

data = clean(data)
test = clean(test)

```

```python
data.head()
```

##Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

y = data['Transported']
x = data.drop('Transported', axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
clf = LogisticRegression(random_state=0, max_iter=1000).fit(x_train, y_train)
```

```python
predictions = clf.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
submission_preds = clf.predict(test)
```

```python
submission = pd.DataFrame({"PassengerId":test_ids.values,
                   "Transported": submission_preds,
                   })
```
